1. How to get mean up-to n number (i.e. n=7) with following condition:

    a) Only for even number up-to n (i.e. mean of 2, 4, 6...n)

    b) Only for odd number up-to n (i.e. mean of 1, 3, 5...n)
    
    c) Only for prime number up-to n (i.e. mean of 2,3,5...n)

```python
def mean_of_even_numbers(n):
    even_numbers = [i for i in range(2, n+1) if i % 2 == 0]
    if even_numbers:
        return sum(even_numbers) / len(even_numbers)
    return 0

def mean_of_odd_numbers(n):
    odd_numbers = [i for i in range(1, n+1) if i % 2 != 0]
    if odd_numbers:
        return sum(odd_numbers) / len(odd_numbers)
    return 0

def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def mean_of_prime_numbers(n):
    prime_numbers = [i for i in range(2, n+1) if is_prime(i)]
    if prime_numbers:
        return sum(prime_numbers) / len(prime_numbers)
    return 0

# Example usage
n = 7
print(f"Mean of even numbers up to {n}: {mean_of_even_numbers(n)}")
print(f"Mean of odd numbers up to {n}: {mean_of_odd_numbers(n)}")
print(f"Mean of prime numbers up to {n}: {mean_of_prime_numbers(n)}")
```

2. Write a Python program to classify a student's grade based on their score:

        a) If the score is 90 or above, print "Grade A"

        b) If the score is between 80 and 89, print "Grade B"

        c) If the score is between 70 and 79, print "Grade C"
    
        d) If the score is below 70, print "Fail"
        

```python
def classify_grade(score):
    if score >= 90:
        return "Grade A"
    elif 80 <= score < 90:
        return "Grade B"
    elif 70 <= score < 80:
        return "Grade C"
    else:
        return "Fail"

# Example usage
score = int(input("Enter the score: "))
print(f"The grade for a score of {score} is: {classify_grade(score)}")
```

3. Write a Python program that accepts a person's age and checks if they are eligible to vote:

    a) If age is 18 or more, print "Eligible to vote".

    b) If age is below 18 but greater than or equal to 12, print "Too young to vote".

    c) If age is below 12, print "Child".
    

```python
def check_voting_eligibility(age):
    if age >= 18:
        return "Eligible to vote"
    elif 12 <= age < 18:
        return "Too young to vote"
    else:
        return "Child"

# Example usage
age = int(input("Enter the age: "))
print(f"Age {age}: {check_voting_eligibility(age)}")
```

1. What is datatype and explain various datatype in brief?

""" A data type is a classification that specifies the type of data that a variable can hold. It determines the kind of operations that can be performed on the data, the way the data is stored, and the range of values that can be represented."""

1. Numeric Types:

-> int: Represents integer values. Example: 1, 42, -7
-> float: Represents floating-point (decimal) values. Example: 3.14, -0.001, 2.0
-> complex: Represents complex numbers with real and imaginary parts. Example: 3+4j, 1-2j



2. Sequence Types:

-> str: Represents a sequence of characters (strings). Example: "hello", 'world'
-> list: Represents an ordered collection of items, which can be of different types. Example: [1, 2, 3], ['a', 'b', 'c']
-> tuple: Similar to lists, but immutable (cannot be changed after creation). Example: (1, 2, 3), ('a', 'b', 'c')



3. Mapping Type:

-> dict: Represents a collection of key-value pairs. Example: {'name': 'Alice', 'age': 25}, {'a': 1, 'b': 2}



4. Set Types:

-> set: Represents an unordered collection of unique items. Example: {1, 2, 3}, {'a', 'b', 'c'}
-> frozenset: An immutable version of a set. Example: frozenset([1, 2, 3])



5. Boolean Type:

-> bool: Represents Boolean values, either True or False. Example: True, False



6. Binary Types:

-> bytes: Represents a sequence of bytes (immutable). Example: b'hello'
-> bytearray: Represents a mutable sequence of bytes. Example: bytearray(b'hello')
-> memoryview: Provides a view of the memory of another binary object.

2. Use the basic function of pandas (like df.dtypes/columns/size/shape/values) to fulfill thefollowing requirement:

    a) How many records in the data frames?

    b) How many elements are there?

    c) What are the names of columns?

    d) What is the datatype of all columns?
    

```python
import pandas as pd

# Create a sample DataFrame
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 35, 40, 28],
    "Score": [85, 90, 88, 92, 95]
}

df = pd.DataFrame(data)

# a) How many records in the data frame?
num_records = df.shape[0]
print(f"Number of records: {num_records}")

# b) How many elements are there?
num_elements = df.size
print(f"Number of elements: {num_elements}")

# c) What are the names of columns?
column_names = df.columns.tolist()
print(f"Column names: {column_names}")

# d) What is the datatype of all columns?
column_dtypes = df.dtypes
print(f"Data types of columns:\n{column_dtypes}")
```

3. Write a Python program to plot a line graph of the following data:

    x = [1, 2, 3, 4, 5]

    y = [2, 4, 6, 8, 10] 
    
    Label the axes as "X-axis" and "Y-axis," and give the title as "BasicLine Plot."

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y, marker='o', linestyle='-', color='b')

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a title
plt.title('Basic Line Plot')

# Show the plot
plt.show()
```

5. Write a Python program to create a pie chart visualizing the distribution of preferences forvarious sports among a group of students.

Use the following details:

Labels: ['Football', 'Basketball', 'Tennis', 'Cricket', 'Swimming']

Sizes: [35, 25, 15, 20, 5]

Colors: ['blue', 'orange', 'green', 'red', 'cyan']

Explode: Explode the slices for "Football," "Basketball," and "Cricket."

Add a shadow for a 3D effect, percentage labels (autopct='%1.1f%%'), and a border for the slices (wedgeprops= {'edgecolor': 'black'}).

The pie chart should:
 Be titled "Sports Preferences Among Students."
 Include a legend in the upper left corner.
 Have a figure size of 10x10.
 Start the pie chart at an angle of 90 degrees for better alignment.

```python
import matplotlib.pyplot as plt

# Data
labels = ['Football', 'Basketball', 'Tennis', 'Cricket', 'Swimming']
sizes = [35, 25, 15, 20, 5]
colors = ['blue', 'orange', 'green', 'red', 'cyan']
explode = (0.1, 0.1, 0, 0.1, 0)  # Explode slices for "Football," "Basketball," and "Cricket"

# Create a pie chart
plt.figure(figsize=(10, 10))  # Set figure size
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=90, shadow=True, wedgeprops={'edgecolor': 'black'})

# Add a title
plt.title('Sports Preferences Among Students')

# Add a legend
plt.legend(labels, title="Sports", loc="upper left")

# Show the plot
plt.show()
```
