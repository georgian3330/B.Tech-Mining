```python
import pandas as pd
import numpy as np

# Create a sample dataset with missing values
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve"],
    "Age": [25, 30, 35, np.nan, 28],
    "Score": [85, 90, np.nan, 88, 95]
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Display the original dataset
print("Original Dataset with missing values: ")
print(df)

# Impute missing values using the mean of the column
df_imputed = df.copy()
df_imputed["Age"] = df_imputed["Age"].fillna(df_imputed["Age"].mean())
df_imputed["Score"] = df_imputed["Score"].fillna(df_imputed["Score"].mean())

# Display the dataset after imputation
print("\nDataset with missing values imputed usign mean: ")
print(df_imputed)
```

```python
import pandas as pd
import numpy as np

# Create a complex dataset with multiple missing values
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Helen"],
    "Income": [50000, np.nan, 75000, 60000, np.nan, 82000, 54000, 91000],
    "Age": [25, 30, 35, np.nan, 28, np.nan, 45, 50],
    "Score": [85, 90, np.nan, 88, 95, 77, np.nan, 89]
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Display the original dataset
print("Original Dataset with missing values: ")
print(df)

# Impute missing values using the mean of the columns
df_imputed = df.copy()
for column in ["Age", "Income", "Score"]:
    df_imputed[column] = df_imputed[column].fillna(df_imputed[column].mean()).round(1)

# Display the dataset after imputation
print("\nDataset with missing values imputed usign mean(Rounded to 1 decimal place): ")
print(df_imputed)
```

```python
import pandas as pd
import numpy as np

# Create a complex dataset with multiple missing values
data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Helen"],
    "Income": [50000, np.nan, 75000, 60000, np.nan, 82000, 54000, 91000],
    "Age": [25, 30, 35, np.nan, 28, np.nan, 45, 50],
    "Score": [85, 90, np.nan, 88, 95, 77, np.nan, 89]
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Display the original dataset
print("Original Dataset with missing values: ")
print(df)

# Impute missing values using the mode of the columns
df_imputed_mode = df.copy()
for column in ["Age", "Income", "Score"]:
    df_imputed_mode[column] = df_imputed_mode[column].fillna(df_imputed_mode[column].mode()[0]).round(1)

# Display the dataset after imputation
print("\nDataset with missing values imputed usign mode(Rounded to 1 decimal place): ")
print(df_imputed_mode)
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import KNNImputer

# Create a dataset with missing values
data = {
    "Age": [25, 30, 35, np.nan, 28, np.nan, 45, 50],
    "Income": [50000, np.nan, 75000, 60000, np.nan, 82000, 54000, 91000],
    "Score": [85, 90, np.nan, 88, 95, 77, np.nan, 89]
}

# Convert to a DataFrame
df = pd.DataFrame(data)

# Display the original dataset
print("Original Dataset with missing values: ")
print(df)

# Create a KNN imputer instance with k = 2 ( 2 nearest neighbors)
imputer = KNNImputer(n_neighbors=2)

# Apply the imputer to the dataset
df_imputed = pd.DataFrame(imputer.fit_transform(df), columns = df.columns)

# Display the dataset after  KNN imputation
print("\nDataset with missing values imputed usign KNN imputer: ")
print(df_imputed)
```
