Create a function to calculate the factorial of a number

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)
print(factorial(3))
```

Factorial

```python
# Example: Function to calculate factorial
def factorial(n):
    if n == 1 or n == 0:
        return 1
    else:
        return n * factorial(n - 1)
# using the function
num = int(input("Enter a number to calculate its factorial: "))
result = factorial(num)
print(f"The factorial of {num} is {result}")
```

Calculator

```python
def calculator(a, b, operator):
    if operator == '+':
        return a + b
    elif operator == '-':
        return a - b
    elif operator == '*':
        return a * b
    elif operator == '/':
        return a / b if b != 0 else "Division by zero error"
    else:
        return "Invalid operator"

# Input from user
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
operator = input("Enter an operator(+, -, *, /): ")

# Function call and result
result = calculator(a, b, operator)
print(f"The result of {a} {operator} {b} is {result}")
```

Temperature Conversion

```python
# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9 / 5 + 32)

celsius_temp = float(input("Enter temperature in celsius: "))
fahrenheit_temp = celsius_to_fahrenheit(celsius_temp)
print(f"{celsius_temp} degrees Celsius is equal to {fahrenheit_temp} degrees Fahrenheit.")
```
