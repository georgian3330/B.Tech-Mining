Generate Random Numbers Between ? and ?

```python
import numpy as np

# Define rand_integers if not already defined
rand_integers = np.random.randint(1, 100, size=3)

min_val = np.min(rand_integers)
max_val = np.max(rand_integers)
rand_arr = min_val + (max_val - min_val) * np.random.rand(3)
print(rand_arr)
```

Create a 3x3 matrix of random numbers between ? and ?

```python
import numpy as np

rand_integers = np.random.randint(1, 10, size = 3)
min_val = np.min(rand_integers)
max_val = np.max(rand_integers)
rand_matrix = min_val + (max_val - min_val) * np.random.rand(3, 3)
print(rand_matrix)
```

Linear Equations

```python
import numpy as np
# Coefficient matrix (A)
A = np.array([[2, 3], [1, -4]])

# Constant vector (B)
B = np.array([8, -2])

# Solve for X 
X = np.linalg.solve(A, B)

print("Solution (x, y): ", X)
```

```python
import numpy as np
temperatures = np.array([25, 27, 24, 23, 22, 26, 28, 30, 29, 25, 23, 24])

# Mean
mean_temp = np.mean(temperatures)

# Median
median_temp = np.median(temperatures)

# Standard Deviation
std_dev_temp = np.std(temperatures)

print("Mean Temperature: ", mean_temp)
print("Median Temperature: ", median_temp)
print("Standard Deviation of Temperature: ", std_dev_temp)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 5x5 grayscale image with random values (0 to 255)
image = np.random.randint(0, 256, (5, 5), dtype = np.uint8)

print("Original Image: \n", image)

# Apply a threshold to highlight brighter regions
threshold = 128
processed_image = np.where(image > threshold, 255, 0)

print("Processed Image (Threshold Applied): \n", processed_image)

# Visualise the image
plt.subplot(1, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.colorbar()

plt.subplot(1, 2, 2)
plt.title("Processed Image")
plt.imshow(processed_image, cmap='gray')
plt.colorbar()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
years = [2015, 2016, 2017, 2018, 2019, 2020]
sales = [250, 300, 400, 350, 450, 500]

# Create the line plot
plt.plot(years, sales, marker='o', color='blue', label='Annual Sales')
plt.title('Annual Sales Over Years')
plt.xlabel('Year')
plt.ylabel('Sales (in $1000)')
plt.grid(True)
plt.legend()
plt.show()
```
